#Homework1 หา Frequent Item Set และ Assosiation Rules
-  ของช่วงเวลา Chrismas & New year (1 Dec-31 Dec) [หนามเตย]
-  ของประเทศในทวีป Europe [จจ]
-  ของช่วงเวลา ตรุษจีนในทวีป Asia(15 Jan-15 Feb) [ใบเตย]
-  ของช่วงเวลา Chrismas & New year ในทวีป Europe (1 Dec-31 Dec) [พิม]
-  ของวันวาเลนไทน์ [ตุนนี่]
-  โดยจัดกลุ่มตามประเทศ+เดือน [พิมจ๊อก]

```python
data
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Filter data for the Christmas and New Year period in Europe
start_date = '2014-12-01'
end_date = '2014-12-31'
european_countries = ['Germany', 'France', 'Spain', 'Italy', 'United Kingdom']

filtered_data = data[(data['InvoiceDate'] >= start_date) & (data['InvoiceDate'] <= end_date) & (data['Country'].isin(european_countries))]

# Drop rows with missing or zero Quantity values
filtered_data = filtered_data.dropna(subset=['Quantity'])
filtered_data = filtered_data[filtered_data['Quantity'] > 0]

# Convert Quantity to integer
filtered_data['Quantity'] = filtered_data['Quantity'].astype(int)

# One-hot encode the data for Apriori algorithm
basket = (filtered_data.groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

# Convert quantity values to 0 or 1
basket_sets = basket.applymap(lambda x: 1 if x > 0 else 0)

# Apply Apriori algorithm to find frequent item sets
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)

# Generate association rules
association_rules_result = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Filter rules for Europe and Christmas/New Year period
filtered_rules = association_rules_result[association_rules_result['consequents'].astype(str).str.contains('Christmas|New Year')]
filtered_rules = filtered_rules[filtered_rules['antecedents'].astype(str).str.contains('Europe')]

# Display the frequent item sets and association rules
print("Frequent Item Sets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(filtered_rules)
```
