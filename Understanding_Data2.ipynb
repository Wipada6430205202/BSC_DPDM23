#Bar chart

#Stacked

```python
import matplotlib.pyplot as plt

# Example shopping mall data
shopping_malls = ['Mall A', 'Mall B', 'Mall C']
electronics = [20, 30, 25]  # Quantities of electronics bought in each mall
clothing = [15, 25, 20]  # Quantities of clothing bought in each mall
groceries = [40, 35, 45]  # Quantities of groceries bought in each mall

# Plotting the stacked bar chart
plt.figure(figsize=(8, 6))

# Bottom bars (electronics)
plt.bar(shopping_malls, electronics, label='Electronics')

# Middle bars (clothing, stacked on top of electronics)
plt.bar(shopping_malls, clothing, bottom=electronics, label='Clothing')

# Top bars (groceries, stacked on top of electronics and clothing)
plt.bar(shopping_malls, groceries, bottom=[sum(x) for x in zip(electronics, clothing)], label='Groceries')

# Adding labels, title, and legend
plt.xlabel('Shopping Malls')
plt.ylabel('Quantities')
plt.title('Stacked Bar Chart of Shopping Mall Purchases')
plt.legend()

plt.show()
```
